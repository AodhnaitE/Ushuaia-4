```python
# Importing Dependencies

import pandas as pd
import os
from sqlalchemy import create_engine
import warnings 
warnings.simplefilter("ignore")
```

```python
# Creating csv path and loading first data in Pandas DataFrame

csv_path = ("data/genres_v2.csv")
song_df = pd.read_csv(csv_path)
song_df.head()
```

```python
# Creating csv path and loading second data in Pandas DataFrame

csv_path = ("data/data_moods.csv")
data_mood = pd.read_csv(csv_path)
data_mood.head()
```

### Data Cleaning and Transformation

```python
# Dropping unwanted columns 

song_df.drop(["uri","track_href","analysis_url","type","title","Unnamed: 0"], axis=1, inplace = True)
```

```python
# Dropping rows with nan values

data_df = song_df.dropna(axis=0)
```

```python
# Converting duration ms to min

data_df['duration_min'] = data_df['duration_ms']/60000
data_df.drop('duration_ms',axis=1,inplace=True)
```

### Using pandas to load csv converted DataFrame into database

```python
# Loading the reported_cases csv dataframe into prosgresql table

data_df.to_sql(name='music_data', con=engine, if_exists='replace', index=False)
data_mood.to_sql(name='data_mood', con=engine, if_exists='replace', index=False)
```
